# Hackman (Hangman using HMMs)

```python
def load_data(file):
  cleaned_words = set()
  with open(file,"r") as file:
    for line in file:
      word = line.strip().upper()
      cleaned_words.add(word)
  return list(cleaned_words)
```

```python
corpus = load_data("corpus.txt")
test = load_data("test.txt")

print("Loaded data")
```
